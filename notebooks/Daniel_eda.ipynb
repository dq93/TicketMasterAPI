```python
ax = datetime_df['ONSALE_START_DATETIME'].value_counts().head(30).plot(kind='bar', title='Top 30 most common sale start dates with hours')
ax.set_xlabel('Dates with hours')
ax.set_ylabel('# of events')
```

January 5 2025 at 8 PM and August 20 2024 at 5 PM were by far the most common times for sales to start. January, October, and November seem to be the most common months.

```python
ax = datetime_df['ONSALE_END_DATETIME'].value_counts().head(30).plot(kind='bar', title='Top 30 most common sale end dates with hours')
ax.set_xlabel('Dates with hours')
ax.set_ylabel('# of events')
```

The sales end dates are far more evenly distributed compared to the sales start dates.

```python
mins = df[df['MAX_PRICE'] <= 500]
ax = mins['MAX_PRICE'].plot(kind='hist', bins=20, title='Prices up to $500')
ax.set_xlabel('Price by USD $')
```

The top most common price for tickets is about $25, followed by $20, and then very surprisingly $0. This could be bad data, but I suspect they are actually free tickets given away by venues in order to inflate attendance numbers for events. I broke up the ticket prices due to extreme outliers. This visualization shows tickets up to $500.

```python
mids = df[(df['MAX_PRICE'] > 500) & (df['MAX_PRICE'] < 5000)]
ax = mids['MAX_PRICE'].plot(kind='hist', bins=20, title='Prices between $500  & $5000')
ax.set_xlabel('Price by USD $')
```

This visualization shows tickets above 500 dollars and under 5000 dollars. The cheaper tickets are more popular but we see a big rise in popularity in tickets around 1600 dollars. This could possibly be the prices for floor seats for many events.

```python
maxs = df[df['MAX_PRICE'] >= 5000]
ax = maxs['MAX_PRICE'].plot(kind='hist', bins=20, title='Top Max Prices')
ax.set_xlabel('Price by USD $')
```

These are the tickets 5000 dollars and up. There seem to be 2 extreme outliers which made the visualizations for all prices hard to see, and that is the highest price for all tickets in the dataset, a single ticket for $$21,000. There is also another ticket for nearly $18,000. Besides those 2, no other tickets even reach $$14,000.

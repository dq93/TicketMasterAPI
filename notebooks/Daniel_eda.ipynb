```python
# creates a separate df for the datetime that drops rows with NaN values and then resets the index
datetime_df = df.dropna().reset_index(drop=True)
datetime_df.shape
```

```python
# convert columns to datetime objects and keep only the date and hours
datetime_df['ONSALE_START_DATETIME'] = pd.to_datetime(datetime_df['ONSALE_START_DATETIME']).dt.floor('h')
datetime_df['ONSALE_END_DATETIME'] = pd.to_datetime(datetime_df['ONSALE_END_DATETIME']).dt.floor('h')

monthdf = datetime_df.copy()

monthdf['start year month'] = datetime_df['ONSALE_START_DATETIME'].dt.to_period('M')
monthdf['end year month'] = datetime_df['ONSALE_END_DATETIME'].dt.to_period('M')


grouped_months = monthdf.groupby('end year month').size().reset_index()

grouped_months
```

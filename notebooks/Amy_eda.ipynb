```python
# STEP 3: Data Preparation

df = df.loc[~df.duplicated(subset=['EVENT_START_DATETIME', 'PROMOTER_NAME', 'MIN_PRICE', 'MAX_PRICE'])] \
    .reset_index(drop=True).copy()

df.head(10)
```

```python
df = df.dropna()
```

```python
earliest_date = df['EVENT_START_DATETIME'].min()
print("Earliest Date:", earliest_date)
```

```python
latest_date = df['EVENT_START_DATETIME'].max()
print("Latest Date:", latest_date)
```

```python
df.shape
```

```python
# single variable analyis (univariate)

ax = df['PROMOTER_NAME'].value_counts() \
    .head(10) \
.plot(kind= 'barh', title= 'Total Events By Promoter in Fall 2024-Fall 2026')
ax.set_xlabel('Count')
ax.set_ylabel('Name of Promoter')
plt.show()
```

```python
avg_ticket_prices = df.groupby('PROMOTER_NAME')['MAX_PRICE'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='MAX_PRICE', y='PROMOTER_NAME', data=avg_ticket_prices)
plt.xlim(0,1000)
plt.ylim(1,10)
plt.xticks(rotation=45)  
plt.title('Average Ticket Price by Promoter')
plt.ylabel('Promoter')
plt.xlabel('Average Ticket Price')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='MAX_PRICE', y='PROMOTER_NAME', data=df)
plt.ylim(1,10)
plt.xlim(0,20000)
plt.xticks(rotation=45)
plt.title('Ticket Price Distribution by Promoter (by Max. Prices)')
plt.xlabel('Max. Prices ($)')
plt.ylabel('Promoter')
plt.subplots_adjust(left=0.1, right=1.0, top=1.0, bottom=0.1)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='MIN_PRICE', y='PROMOTER_NAME', data=df)
plt.ylim(1,10)
plt.xlim(0,1000)
plt.xticks(rotation=45)
plt.title('Ticket Price Distribution by Promoter (by Min. Price)')
plt.xlabel('Min. Prices ($)')
plt.ylabel('Promoter')
plt.subplots_adjust(left=0.90, right=1.0, top=1.0, bottom=0.90)
plt.tight_layout()
plt.show()
```

```python
df['EVENT_START_DATETIME'] = pd.to_datetime(df['EVENT_START_DATETIME'])
df['YEAR'] = pd.to_datetime(df['EVENT_START_DATETIME']).dt.year

avg_price_by_year = df.groupby('YEAR')[['MAX_PRICE', 'MIN_PRICE']].mean()
avg_price_by_year.plot(kind='line', figsize=(10, 6))

plt.xticks(avg_price_by_year.index.astype(int))
plt.title('Average Ticket Price Over Time')
plt.xlabel('Year')
plt.ylabel('Average Ticket Price')
plt.show()
```

## My Analysis

10 promoters were compared based on several factors: Total Events, Average Ticket Price, and Ticket Price Distribution.

### Key Findings

The NBA Regular Season is one of the top performers out of all the promoters that were analyzed. 

All Elite Wrestling had one of the highest average ticket prices. Seattle Theater Group's average ticket price was significantly lower. Prices seem to average higher for sports-related promoters, and they average lower for other types of promoters like theater or live musical performances.

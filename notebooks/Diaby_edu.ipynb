Number of Events | Music vs. Sport

```python
# Bar graph for Number of Events by Classification (Music vs. Sport)

counts = df["Genre"].value_counts()
counts.plot(kind="bar")
plt.xlabel("genre")
plt.xticks(rotation=90)
plt.ylabel("Number of Events")
plt.show()
```

```python
# Future  max prices related to their Sub_genre
plt.figure(figsize=(12, 6))
sns.boxplot(x='Max_Price', y='Sub_Genre', data=df)
plt.ylim(1,10)
plt.xlim(0,20000)
plt.xticks(rotation=45)
plt.title('Ticket Price Distribution by Sub_Genre (by Max. Prices)')
plt.xlabel('Max. Prices ($)')
plt.ylabel('Sub_Genre')
plt.subplots_adjust(left=0.1, right=1.0, top=1.0, bottom=0.1)
plt.tight_layout()
plt.show()
```

Pop and Country are on their of becoming the most expensive shows.
